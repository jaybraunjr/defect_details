```python


def save_defect_data_to_json(defects_per_frame, base_name, output_dir='output'):
    """
    Save only the defect count per frame to a JSON file.
    
    :param defects_per_frame: List of defects per frame
    :param base_name: Base name for the output file
    :param output_dir: Directory to save the JSON file
    """
    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Structure the JSON output
    defect_data = {}
    for frame_data in defects_per_frame:
        frame_num = frame_data['frame']  # Get the frame number
        
        # Ensure defects are correctly retrieved and saved
        defect_count = frame_data['defect_count']  # Extract defect count
        
        # Save the defect count for this frame
        defect_data[f"frame_{frame_num}"] = {
            "defects_count": defect_count  # Just store the number of defects
        }
    
    # Save to a JSON file
    output_file = os.path.join(output_dir, f'{base_name}_defects_count.json')
    with open(output_file, 'w') as f:
        json.dump(defect_data, f, indent=4)
    
    print(f"Defect count data saved to {output_file}")

```

```python
def calculate_average_and_std(base_name, output_dir='output', n_blocks=6):

    input_file = os.path.join(output_dir, f'{base_name}_defects_count.json')
    with open(input_file, 'r') as f:
        defect_data = json.load(f)
    
    # Extract the defect counts for all frames
    defect_counts = [frame_data['defects_count'] for frame_data in defect_data.values()]
    n_frames = len(defect_counts)
    
    # Ensure the number of frames is divisible by the number of blocks
    block_size = n_frames // n_blocks
    if n_frames % n_blocks != 0:
        print(f"Warning: Number of frames ({n_frames}) is not evenly divisible by {n_blocks} blocks. Using only full blocks.")
        defect_counts = defect_counts[:block_size * n_blocks]  # Truncate to fit the full blocks

    # Reshape defect counts into blocks and calculate the average defect count per block
    block_averages = np.mean(np.array(defect_counts).reshape(n_blocks, block_size), axis=1)
    
    # Calculate the overall average and standard deviation of the block-averaged data
    average_defects = np.mean(block_averages)
    std_defects = np.std(block_averages)
    
    # Save the summary statistics to a new JSON file
    summary_data = {
        "average_defects_per_frame": average_defects,
        "std_defects_per_frame": std_defects,
        "block_averages": block_averages.tolist(),  # Save the individual block averages
        "block_size": block_size
    }
    
    # Save the summary to a JSON file
    summary_output_file = os.path.join(output_dir, f'{base_name}_defects_summary_block_averaged.json')
    with open(summary_output_file, 'w') as f:
        json.dump(summary_data, f, indent=4)
    
    print(f"Summary data (block-averaged) saved to {summary_output_file}")
```

```python
defect_types = ['PLacyl', 'TGacyl', 'TGglyc']

for defect_type in defect_types:
    calculate_average_and_std(base_name=f'test_{defect_type}', output_dir='output_drude')
```

```python
import matplotlib.cm as cm

def plot_PLacyl_defects_across_outputs_block_averaged(
    output_dirs,
    output_labels,
    defect_type='PLacyl',
    y_limit=None,
    cmap='Greens',  # Colormap for gradient shades
    figsize=(4, 4),
    error_bars='std',
    capsize=5,
    linewidth=1.5,
    bar_width=0.6,
    alpha=0.8,
    title='Defect frequency'
):

    averages = []
    errors = []

    # Load data from JSON files
    for output_dir in output_dirs:
        summary_file = os.path.join(output_dir, f'test_{defect_type}_defects_summary_block_averaged.json')
        with open(summary_file, 'r') as f:
            summary_data = json.load(f)
            averages.append(summary_data['average_defects_per_frame'])
            if error_bars == 'std':
                errors.append(summary_data['std_defects_per_frame'])  # Standard deviation
            elif error_bars == 'sem':
                errors.append(summary_data['std_defects_per_frame'] / np.sqrt(4))  # SEM calculation

    fig, ax = plt.subplots(figsize=figsize)
    ax.set_facecolor('white')  # Set a clean white background
    x_pos = np.arange(len(output_labels))

    # Generate a colormap based on the number of bars
    colors = cm.get_cmap(cmap)(np.linspace(0.3, 1, len(output_labels)))  # Colormap gradient

    # Plot bars with gradient shades of green
    bars = ax.bar(
        x_pos,
        averages,
        yerr=errors,
        align='center',
        alpha=alpha,
        ecolor='black',
        capsize=capsize,
        color=colors,
        edgecolor='black',
        linewidth=linewidth,
        width=bar_width
    )

    # Add grid - light grey and only horizontal
    ax.grid(True, which='major', axis='y', linestyle='--', linewidth=0.8, color='grey', alpha=0.7)

    # Set labels and title with professional fonts
    ax.set_ylabel('Frequency per frame', fontsize=11, fontweight='bold')
    ax.set_title(title, fontsize=12, fontweight='bold', pad=15)
    ax.set_xticks(x_pos)
    ax.set_xticklabels(output_labels, fontsize=12, rotation=45, ha='right')

    # Handle y-limit range
    if isinstance(y_limit, (tuple, list)) and len(y_limit) == 2:
        ax.set_ylim(y_limit[0], y_limit[1])  # Use both min and max
    elif y_limit:
        ax.set_ylim(0, y_limit)  # Use y_limit as the maximum

    # Adjust y-axis tick labels to be larger and bold
    ax.tick_params(axis='y', labelsize=12)
    for label in ax.get_yticklabels():
        label.set_fontweight('bold')

    # Enhance layout
    plt.tight_layout()

    # Save the figure with high resolution
    plt.savefig('freq_PLacyl_50_professional.png', dpi=500, bbox_inches='tight')
    plt.show()

# Example usage
if __name__ == "__main__":
    # Directories and labels for the outputs you want to compare
    output_dirs = [
        'output_drude/',
        'output_c36/',
        'output_c36c/',
        'output_c36c_st10/'
    ]
    output_labels = ['Drude', 'C36s', 'C36c', 'C36c 10 mN/m']

    # Call the plotting function for PLacyl using block-averaged data with gradient shades of green
    plot_PLacyl_defects_across_outputs_block_averaged(
        output_dirs,
        output_labels,
        defect_type='PLacyl',
        y_limit=[20, 43],  # Specify both minimum and maximum y-limit
        alpha=0.8,  # Slightly increased transparency
    )
```

```python
import numpy as np
import os
import json
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

def plot_TGacyl_TGglyc_defects_across_outputs_block_averaged(
    output_dirs,
    output_labels,
    defect_types=['TGacyl', 'TGglyc'],
    y_limit=None,
    colors=None,
    figsize=(3.5,4),
    error_bars='std',
    capsize=5,
    linewidth=1.5,
    bar_width=0.35,
    alpha=0.85,
    title='TGacyl & TGglyc per frame'
):

    averages = {defect_type: [] for defect_type in defect_types}
    errors = {defect_type: [] for defect_type in defect_types}

    # Define a professional color palette if not provided
    if colors is None:
        colors = [
            'red',  # Blue for TGacyl
            'blue'   # Orange for TGglyc
        ]
    else:
        if len(colors) < len(defect_types):
            raise ValueError("Number of colors provided is less than the number of defect types.")

    # Collect data
    for defect_type in defect_types:
        for output_dir in output_dirs:
            summary_file = os.path.join(output_dir, f'test_{defect_type}_defects_summary_block_averaged.json')
            if not os.path.isfile(summary_file):
                raise FileNotFoundError(f"Summary file not found: {summary_file}")
            with open(summary_file, 'r') as f:
                summary_data = json.load(f)
                averages[defect_type].append(summary_data['average_defects_per_frame'])
                if error_bars == 'std':
                    errors[defect_type].append(summary_data['std_defects_per_frame'])  # Standard deviation
                elif error_bars == 'sem':
                    errors[defect_type].append(summary_data['std_defects_per_frame'] / np.sqrt(4))  # SEM calculation
                else:
                    raise ValueError("error_bars parameter must be either 'std' or 'sem'.")

    fig, ax = plt.subplots(figsize=figsize)
    ax.set_facecolor('white')  # Set a clean white background
    x = np.arange(len(output_labels))  # the label locations

    total_defects = len(defect_types)
    # Calculate offsets for grouped bars
    offsets = np.linspace(-bar_width/2, bar_width/2, total_defects)

    # Plot each defect type
    bars = []
    for idx, defect_type in enumerate(defect_types):
        x_pos = x + offsets[idx]
        bar = ax.bar(
            x_pos,
            averages[defect_type],
            yerr=errors[defect_type],
            align='center',
            alpha=alpha,
            ecolor='black',
            capsize=capsize,
            color=colors[idx],
            edgecolor='black',
            linewidth=linewidth,
            width=bar_width,
            label=defect_type
        )
        bars.append(bar)

    # Add a light grey grid on y-axis only
    ax.grid(True, which='major', axis='y', linestyle='--', linewidth=0.75, color='lightgrey', alpha=0.6)

    # Set labels and title with professional fonts
    ax.set_ylabel('Frequency per frame', fontsize=12, fontweight='bold')
    ax.set_title(title, fontsize=14, fontweight='bold', pad=15)
    ax.set_xticks(x)
    ax.set_xticklabels(output_labels, fontsize=11, rotation=45, ha='right')

    if y_limit:
        ax.set_ylim(0, y_limit)

    # Adjust y-axis tick labels to be larger and bold
    ax.tick_params(axis='y', labelsize=12)
    for label in ax.get_yticklabels():
        label.set_fontweight('bold')

    # Remove the legend box and make the legend more professional
    ax.legend(fontsize=12, frameon=False)

    # Enhance layout for a professional appearance
    plt.tight_layout()

    # Save the figure with high resolution
#     plt.savefig('freq_TG_new_professional.png', dpi=600, bbox_inches='tight')
    plt.show()

# Example usage
if __name__ == "__main__":
    output_dirs = [
        'output_drude',
        'output_c36',
        'output_c36c',
        'output_c36c_st10'
    ]
    output_labels = [ 'Drude','C36s', 'C36c','C36c 10 mN/m']

    # Call the plotting function for TGacyl and TGglyc using block-averaged data with enhanced aesthetics
    plot_TGacyl_TGglyc_defects_across_outputs_block_averaged(
        output_dirs=output_dirs,
        output_labels=output_labels,
        defect_types=['TGacyl', 'TGglyc'],
        y_limit=8,
        alpha=0.85,  # Slightly higher opacity for more solid colors
        title='Defect frequencies'
    )
```

```python
def plot_lifetime_with_error(avg_lifetime_data):
    defect_types = avg_lifetime_data.keys()
    means = [avg_lifetime_data[defect_type]['mean'] for defect_type in defect_types]
    errors = [avg_lifetime_data[defect_type]['sem'] for defect_type in defect_types]  # You can use 'std' for standard deviation

    plt.bar(defect_types, means, yerr=errors, capsize=5, color='skyblue')
    plt.ylabel('Average Lifetime (ps)')
    plt.title('Average Defect Lifetime with Error Bars')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
from defect_analysis import calc_probs_for_chunks, plot_all_histograms
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patheffects as patheffects  # Import for text path effects
import matplotlib.ticker as ticker  # Import for tick formatting

def plot_probabilities(datasets, size_thresholds):
    # Mapping from dataset names to display names
    name_mapping = {
        'PLacyl_c36': 'PL-acyl',
        'TGacyl_c36': 'TG-acyl',
        'TGglyc_c36': 'TG-glyc',
        'PLacyl_c36c': 'PL-acyl',
        'TGacyl_c36c': 'TG-acyl',
        'TGglyc_c36c': 'TG-glyc',
        'PLacyl_drude_test': 'PL-acyl',
        'TGacyl_drude_test': 'TG-acyl',
        'TGglyc_drude_test': 'TG-glyc',
        'PLacyl_c36c_st10': 'PL-acyl',
        'TGacyl_c36c_st10': 'TG-acyl',
        'TGglyc_c36c_st10': 'TG-glyc',
    }
    
    # Colors mapped to original dataset names (which are unique)
    colors = {
        'PLacyl_c36': 'green',
        'TGacyl_c36': 'green',
        'TGglyc_c36': 'green',
        'PLacyl_c36c': 'purple',
        'TGacyl_c36c': 'purple',
        'TGglyc_c36c': 'purple',
        'PLacyl_drude_test': 'orange',
        'TGacyl_drude_test': 'orange',
        'TGglyc_drude_test': 'orange',
        'PLacyl_c36c_st10': 'red',
        'TGacyl_c36c_st10': 'red',
        'TGglyc_c36c_st10': 'red',
    }

    # Mapping from dataset names to group labels
    dataset_to_group = {
        'PLacyl_c36': 'C36',
        'TGacyl_c36': 'C36',
        'TGglyc_c36': 'C36',
        'PLacyl_c36c': 'C36c',
        'TGacyl_c36c': 'C36c',
        'TGglyc_c36c': 'C36c',
        'PLacyl_drude_test': 'Drude',
        'TGacyl_drude_test': 'Drude',
        'TGglyc_drude_test': 'Drude',
        'PLacyl_c36c_st10': 'C36c (10 mN/m)',
        'TGacyl_c36c_st10': 'C36c (10 mN/m)',
        'TGglyc_c36c_st10': 'C36c (10 mN/m)',
    }

    group_labels = ['C36', 'C36c', 'Drude', 'C36c (10 mN/m)']
    group_separators = [3, 6, 9]  # Positions to add vertical separators between groups

    # Mapping from group labels to background colors
    group_colors = {
        'C36': 'lightblue',
        'C36c': 'lightgreen',
        'Drude': 'lightyellow',
        'C36c (10 mN/m)': 'lightpink'
    }

    n_plots = len(size_thresholds)
    fig, axes = plt.subplots(n_plots, 1, figsize=(6, 1.8 * n_plots), sharex=True)

    if n_plots == 1:
        axes = [axes]  # Ensure axes is a list even if there's only one subplot

    # Prepare x_positions and labels
    labels = []
    colors_list = []
    for idx_dataset, (name, _) in enumerate(datasets):
        display_name = name_mapping.get(name, name)
        labels.append(display_name)
        # Use the original dataset name to get the color
        colors_list.append(colors.get(name, 'gray'))

    x_positions = np.arange(len(labels))

    for idx_plot, (threshold, ax) in enumerate(zip(size_thresholds, axes)):
        means, std_devs = [], []

        for idx_dataset, (name, (bin_centers_chunks, probabilities_chunks)) in enumerate(datasets):
            mean_prob, std_prob = calc_probs_for_chunks(threshold, bin_centers_chunks, probabilities_chunks)
            print(f"{name_mapping.get(name, name)} - Threshold {threshold}: Mean={mean_prob}, Std={std_prob}")
            means.append(mean_prob)
            std_devs.append(std_prob)

        # Plot bars with thicker black edges
        bars = ax.bar(
            x_positions,
            means,
            yerr=std_devs,
            alpha=0.9,
            color=colors_list,
            edgecolor='black',
            linewidth=2.5,  # Thicker black lines around bars
            capsize=5
        )
        # Set y-axis label with angstrom squared symbol
        ax.set_ylabel(rf'> {threshold} $\mathrm{{\AA^2}}$', fontsize=11, weight='bold')
        ax.tick_params(axis='y', labelsize=8)

        # Format y-axis tick labels to one decimal place
        ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

        # Add shaded regions between the groups
        for i in range(len(group_separators) + 1):
            start = group_separators[i - 1] if i > 0 else 0
            end = group_separators[i] if i < len(group_separators) else len(labels)
            group_label = group_labels[i]
            group_color = group_colors.get(group_label, '#f0f0f0')
            ax.axvspan(start - 0.5, end - 0.5, color=group_color, alpha=0.2)
        
        # Add vertical lines to separate groups
        for sep in group_separators:
            ax.axvline(sep - 0.5, color='black', linestyle='--', lw=1)

        if idx_plot < n_plots - 1:
            ax.set_xticklabels([])
        else:
            ax.set_xlabel('Datasets', fontsize=12)
            ax.set_xticks(x_positions)
            ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=10)

            # Color the x-axis labels and add black outline
            xtick_labels = ax.get_xticklabels()
            for tick_label, name in zip(xtick_labels, [dataset[0] for dataset in datasets]):
                tick_label.set_color(colors.get(name, 'black'))
                # Add black outline to the labels
                tick_label.set_path_effects([
                    patheffects.Stroke(linewidth=1, foreground='black'),
                    patheffects.Normal()
                ])

        # Adjust x-axis limits
        ax.set_xlim(-0.5, len(labels) - 0.5)

    # Place group labels above the plot
    # Calculate the midpoints for each group based on the x_positions
    total_width = len(labels) - 1  # Adjust for zero-based indexing
    group_positions = []
    for i in range(len(group_separators) + 1):
        start = group_separators[i - 1] if i > 0 else 0
        end = group_separators[i] if i < len(group_separators) else len(labels)
        mid_point = (start + end - 1) / 2 / total_width  # Normalize to [0,1]
        group_positions.append(mid_point)

    # Add group labels above the plot
    for pos, label in zip(group_positions, group_labels):
        fig.text(pos, 0.98, label, ha='center', va='bottom', fontsize=10, fontweight='bold', transform=fig.transFigure)

    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to make room for group labels
#     plt.savefig('defects_size2_with_group_labels.png', dpi=600)
    plt.show()

# Generate histograms with block averaging and error bars
histograms = plot_all_histograms(
    defects_PLacyl_drude_test,
    defects_TGacyl_drude_test,
    defects_TGglyc_drude_test,
    defects_PLacyl_c36c_st10,
    defects_TGacyl_c36c_st10,
    defects_TGglyc_c36c_st10,
    n_chunks=6  # Number of chunks for block averaging
)

datasets = [
    ('PLacyl_drude_test', histograms[0]),
    ('TGacyl_drude_test', histograms[1]),
    ('TGglyc_drude_test', histograms[2]),
    ('PLacyl_c36c_st10', histograms[3]),
    ('TGacyl_c36c_st10', histograms[4]),
    ('TGglyc_c36c_st10', histograms[5]),
]

size_thresholds = [15, 50, 75]
plot_probabilities(datasets, size_thresholds)
```
