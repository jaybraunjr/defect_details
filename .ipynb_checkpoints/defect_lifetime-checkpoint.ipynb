```python
def plot_average_lifetime(base_names, min_defect_sizes, output_dir='output'):
    all_averages = {}
    base_name_colors = {
        'drude_c36': 'yellow',
        'c36c': 'green',
        'test': 'orange',
        'v_st10': 'red'
    }
    base_name_labels = {
        'drude_c36': 'C36',
        'c36c': 'C36c',
        'test': 'Drude (eq)',
        'v_st10': 'C36c 10 mN/m'
    }

    for min_size in min_defect_sizes:
        for base_name in base_names:
            file_path = os.path.join(output_dir, f'average_lifetime_{base_name}_minsize{min_size}.json')
            if os.path.exists(file_path):
                with open(file_path, 'r') as f:
                    average_lifetime_per_type = json.load(f)
                    all_averages[f"{base_name}_minsize{min_size}"] = average_lifetime_per_type
            else:
                print(f"File not found: {file_path}, skipping this base name.")

    if not all_averages:
        print("No data found to plot.")
        return

    defect_types = next(iter(all_averages.values())).keys()
    n_plots = len(defect_types)
    y_limit = 6
    fig, axes = plt.subplots(n_plots, 1, figsize=(6, 2.5 * n_plots), sharex=True)

    for ax, defect_type in zip(axes, defect_types):
        defect_type_averages = {key: all_averages[key][defect_type] for key in all_averages.keys()}
        sorted_keys = sorted(defect_type_averages.keys(), key=lambda x: (base_names.index(x.split('_minsize')[0]), int(x.split('minsize')[1])))
        sorted_values = [defect_type_averages[key]['mean'] for key in sorted_keys]
        errors = [defect_type_averages[key]['sem'] * 100 for key in sorted_keys]
        colors = [base_name_colors[key.split('_minsize')[0]] for key in sorted_keys]

        for i, (key, value) in enumerate(zip(sorted_keys, sorted_values)):
            ax.scatter(i, value * 100, color=colors[i], edgecolor='black', linewidth=2, label=key if i == 0 else "", s=130, marker='s')

        ax.set_ylabel('Lifetime (ps)', size=14)
        ax.set_title(f'{defect_type} Defects', size=15, fontweight='bold')
        ax.set_ylim(0, y_limit * 100)
        for y in range(100, y_limit * 100, 100):
            ax.axhline(y, color='gray', linestyle='--', linewidth=1.2)

        ax2 = ax.twinx()
        ax2.scatter(range(len(sorted_values)), errors, color='red', edgecolor='black', linewidth=1.5, label='SEM', s=150, marker='1')
        ax2.set_ylabel('SEM', size=14, color='red')
        ax2.set_ylim(0, max(errors) * 1.5)
        ax2.tick_params(axis='y', labelsize=10, colors='red')
        ax2.spines['right'].set_color('red')

    for ax in axes[:-1]:
        ax.set_xticklabels([])

    num_base_names = len(base_names)
    repeated_labels = [str(size) for size in min_defect_sizes] * num_base_names
    x_ticks = range(len(repeated_labels))

    axes[-1].set_xticks(x_ticks)
    axes[-1].set_xticklabels(repeated_labels, rotation=45, ha='right', size=12.9, fontweight='bold')
    axes[-1].set_xlabel('Min Defect Size ($\AA^2$)', size=15)

    handles = [plt.Line2D([0], [0], marker='s', color='w', markerfacecolor=color, markeredgecolor='black', markersize=10, linestyle='') for color in base_name_colors.values()]
    labels = [base_name_labels[key] for key in base_name_colors.keys()]

    legend = fig.legend(handles, labels, loc='center left', bbox_to_anchor=(1, 0.5), title='systems', fontsize=15)
    legend.get_frame().set_facecolor('white')

    plt.tight_layout()
    plt.show()

base_names = ['drude_c36', 'c36c', 'test', 'v_st10']
min_defect_sizes = [5, 15, 50]

plot_average_lifetime(base_names, min_defect_sizes)
```
